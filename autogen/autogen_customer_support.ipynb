```python
import autogen
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# function to process the url and api 
def read_data_from_url(url):
    try:
        # get request 
        response = requests.get(url)
        response.raise_for_status() 
        soup = BeautifulSoup(response.content, 'html.parser')
        # extract and return the content 
        text = soup.get_text()
        return text
    except Exception as e:
        return str(e)

def read_data_from_api(api_url):
    try:
        # get API 
        response = requests.get(api_url)
        response.raise_for_status()
        # return json content 
        data = response.json()
        return data
    except Exception as e:
        return str(e)
    
# retrieve
def read_data(agent, resource_type, resource_location):
    # Call the function to read data from a URL
    if resource_type == 'url':
        return read_data_from_url(resource_location)
    # Call the function to read data from an API
    elif resource_type == 'api':
        return read_data_from_api(resource_location)
    else:
        return "Unsupported resource type."
```

```python
# agent to read data
data_sources_reader = autogen.ConversableAgent(
    name="data_sources_reader",
    system_message="Identify and retrieve data from various data sources such as APIs, databases, and CSV files.",
    description="Data Sources Reader. Gathers data from multiple sources.",
    llm_config=llm_config,
)
```

```python
# Read data 

# data type
resource_type = 'url'  
# data location 
resource_location = 'https://crewai.com'  
data = read_data(data_sources_reader, resource_type, resource_location)

print(data)
```

```python
# Agent 

# user_proxy agent 
user_proxy = autogen.ConversableAgent(
    name="user_proxy",
    system_message="Receive and understand the customer's inquiry, then relay it to the support agent in clear, concise language. Ensure any questions for the customer are simple and easy to understand, focusing on gathering all necessary information for accurate support.",
    code_execution_config=False,
    llm_config=llm_config,
    human_input_mode="ALWAYS"
)

# Support agent
support_agent = autogen.ConversableAgent(
    name="support_agent",
    system_message=""" 
        You are a Senior Support Representative working at crewAI (https://crewai.com).
        Your primary responsibility is to provide outstanding support to customer, a highly valued client.
        Your approach should be friendly, complete, and fully helpful, leaving no questions unanswered.
        Make no assumptions; respond with clear, comprehensive answers to each inquiry.
    """,
    description="Support agent tasked with providing thorough and helpful answers to customer inquiries.",
    llm_config=llm_config
)

# Quality assurance agent
support_quality_assurance_agent = autogen.ConversableAgent(
    name="support_quality_assurance_agent",
    system_message=""" 
        You are a Support Quality Assurance Specialist at crewAI (https://crewai.com), ensuring top-quality support is given to customer.
        Your task is to review responses from the support representative to confirm they are complete, accurate, and friendly.
        Verify that each aspect of the inquiry is addressed, referencing all necessary sources, and upholding high standards.
        Aim for helpfulness and professionalism, without any assumptions in the response.
    """,
    description="Quality assurance agent responsible for reviewing and ensuring the support quality.",
    llm_config=llm_config
)


final_answer = autogen.ConversableAgent(
    name="final_answer",
    system_message="Give the final output after the all agent do their task and make the generated output neater and more like a blog and remove unused character",
    description="final_answer. Agent that give the final output after all agent do their task",
    llm_config=llm_config
)
```

```python
# group chat 
groupchat = autogen.GroupChat(
    agents=[user_proxy,support_agent, support_quality_assurance_agent, final_answer],
    messages=[],
    max_round=10
)

# Manager agent that manages the group chat
manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config
)
```

```python
task = f"""
        Answer this question based on {resource_location}given : 
        I need help with setting up a Crew and kicking it off, specifically how can I add memory to my crew? Can you provide guidance?
        Give the final answer in neat way
"""
```

```python
# Start 
groupchat_result = user_proxy.initiate_chat(
    manager,
    message=task,
    additional_data={"data": data}
)
```
